# Longest Consecutive Sequence

Given an unsorted array of integers nums, return the length of the longest consecutive elements sequence.

You must write an algorithm that runs in O(n) time.


Example 1:
```
Input: nums = [100,4,200,1,3,2]
Output: 4
Explanation: The longest consecutive elements sequence is [1, 2, 3, 4]. Therefore its length is 4.
```

Example 2:

```
Input: nums = [0,3,7,2,5,8,4,6,0,1]
Output: 9
```

Example 3:

```
Input: nums = [1,0,1,2]
Output: 3
```

Constraints:

```
0 <= nums.length <= 105
-109 <= nums[i] <= 109
```

````python
tests = [
	{
		"nums": [100,4,200,1,3,2],
		"expectedOutput": 4
	},
	{
		"nums": [0,3,7,2,5,8,4,6,0,1],
		"expectedOutput": 9
	},
	{
		"nums": [1,0,1,2],
		"expectedOutput": 3
	},
]
````

````python
def divide(list):
    pivot = list[0]
    leftSide = []
    rightSide = []
    for i in range(1, len(list)):
        if list[i] >= pivot:
            rightSide.append(list[i])
        else:
            leftSide.append(list[i])
    return leftSide, [pivot], rightSide

def quickSort(list):
    if len(list) < 2:
        return list
    else:
        leftSide, pivot, rightSide = divide(list)
    return quickSort(leftSide) + pivot + quickSort(rightSide)

def longest_consecutive_sequence(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    """
    sort (quick sort) / .sort O(n log n)
    contiguos = abs(x - y) = 1
        count += 1
    Time: O()
    Space: O()
    Could work but doesn't achieve Time and Space constraints
    """
    if len(nums) < 1:
        return 0
    sortedList = quickSort(nums)
    topCounter = 0
    counter = 0
    for i in range(0, len(sortedList) - 1):
        if sortedList[i] == sortedList[i + 1]:
            continue
        if abs(sortedList[i] - sortedList[i + 1]) == 1:
            counter += 1
            if counter > topCounter:
                topCounter = counter
        else:
            counter = 0
    return topCounter + 1
````

````python
def longest_consecutive_sequence(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    """
    use a set to remove duplicated numbers
    count streaks only if currentNum - 1 is not in the set (this means it could be the start of a streak)
    then count streaks by iterating the set looking for the consecutive number (currentNum + 1) in the set
    if it exist increase the currentStreak if not compare with the longestStreak and update if higher
    return longestStreak
    Time: O(n)
    Space: O(n)
    """
    nums_set = set(nums)
    longest_streak = 0

    for num in nums_set:
        if num - 1 not in nums_set:
            current_num = num
            current_streak = 1
            while current_num + 1 in nums_set: #Max O(2n)
                current_num += 1
                current_streak += 1
            longest_streak = max(longest_streak, current_streak)
    return longest_streak
````
