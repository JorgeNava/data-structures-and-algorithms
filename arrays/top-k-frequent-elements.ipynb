# Top K Frequent Elements
Given an integer array nums and an integer k, return the k most frequent elements. You may return the answer in any order.

Example 1:

```
Input: nums = [1,1,1,2,2,3], k = 2
Output: [1,2]
```

Example 2:

```
Input: nums = [1], k = 1
Output: [1]
```

Example 3:

```
Input: nums = [1,2,1,2,1,2,3,1,3,2], k = 2
Output: [1,2]
```

Constraints:

- 1 <= nums.length <= 10^5
- -10^4 <= nums[i] <= 10^4
- k is in the range [1, the number of unique elements in the array].
- It is guaranteed that the answer is unique.

Follow up: Your algorithm's time complexity must be better than O(n log n), where n is the array's size.

````python
tests = [
	{
		"nums": [1,1,1,2,2,3],
		"k": 2,
		"expectedOutput": [1,2]
	},
	{
		"nums": [1],
		"k": 1,
		"expectedOutput": [1]
	},
	{
		"nums":  [1,2,1,2,1,2,3,1,3,2],
		"k": 2,
		"expectedOutput": [1,2]
	},
]
````

````python
def top_k_frequent_elements(nums, k):
    """
    :type nums: List[int]
    :type k: int
    :rtype: List[int]
    """
    """
    Go through the array and if the number is in our dict then add 1 if not set it to 1
    When done, go through the dict K times and obtain their top numbers comparing one with another
    and storing the biggest one on and array and then removing it from the dict 
    Time: O(k * n), worst-case scenario O(n^2)
    Space: O(n)
    """
    counter = dict()
    for num in nums:
        if num not in counter:
            counter[num] = 1
        else:
            counter[num] += 1
    most_frequents = []
    for i in range(0, k):
        most_frequency = next(iter(counter))
        higher_freq_num = next(iter(counter.values()))
        for unique_num, unique_num_freq in counter.items():
            if unique_num_freq > higher_freq_num:
                higher_freq_num = unique_num_freq
                most_frequency = unique_num
        most_frequents.append(most_frequency)
        counter.pop(most_frequency)
    return most_frequents
````

````python
def top_k_frequent_elements(nums, k):
    """
    :type nums: List[int]
    :type k: int
    :rtype: List[int]
    """
    count = {}
    for num in nums:
        count[num] = 1 + count.get(num, 0)
    
    freq = [[] for i in range(len(nums) + 1)]

    for n, c in count.items():
        freq[c].append(n)
    
    res = []
    for i in range(len(freq) - 1, 0, -1):
        for n in freq[i]:
            res.append(n)
            if len(res) == k:
                return res
````

````python
for test in tests:
    output = top_k_frequent_elements(test["nums"], test["k"])
    print('output',output)
    print('>', output == test["expectedOutput"])
````
