# Valid Palindrome
A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers.

Given a string s, return true if it is a palindrome, or false otherwise.


Example 1:

Input: s = "A man, a plan, a canal: Panama"
Output: true
Explanation: "amanaplanacanalpanama" is a palindrome.


Example 2:

Input: s = "race a car"
Output: false
Explanation: "raceacar" is not a palindrome.


Example 3:

Input: s = " "
Output: true
Explanation: s is an empty string "" after removing non-alphanumeric characters.
Since an empty string reads the same forward and backward, it is a palindrome.
 

Constraints:

1 <= s.length <= 2 * 10^5
s consists only of printable ASCII characters.

```python
tests = [
	{
		"s": "A man, a plan, a canal: Panama",
		"expectedOutput": True
	},
	{
		"s": "race a car",
		"expectedOutput": False
	},
	{
		"s": " ",
		"expectedOutput": True
	},
]
```

# Solution #1
This algorithm has a time complexity of O(n) since it iterates through every element on the string and a space complexity of O(n) since we recreate strings while doing the cleanup.

```python
import re

def is_palindrome(s):
    """
    :type s: str
    :rtype: bool
    """
    clean_string = re.sub(r'[^a-z0-9]', '', s.lower())
    left, right = 0, len(clean_string) - 1
    for i in range(len(clean_string)):
      if clean_string[left] != clean_string[right]:
        return False
      left += 1
      right -= 1
    return True
```

# Solution #2
This algorithm has a time complexity of O(n) since it iterates through every element on the string and a space complexity of O(1) since we now only use the two left and right pointers.

The difference here is that we remove the cleaning from the string and did conversions and checks to each separated char of the string.

```python
def is_palindrome(s):
    """
    :type s: str
    :rtype: bool
    """
    left, right = 0, len(s) - 1
    while left < right:
      while left < right and not s[left].isalnum():
          left += 1
      while left < right and not s[right].isalnum():
          right -= 1

      if s[left].lower() != s[right].lower():
        return False
      left += 1
      right -= 1
    return True
```

```python
for test in tests:
    output = is_palindrome(test["s"])
    print('output',output)
    print('>', output == test["expectedOutput"])
```
