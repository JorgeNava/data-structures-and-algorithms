# Reverse String
Write a function that reverses a string. The input string is given as an array of characters s.

You must do this by modifying the input array in-place with O(1) extra memory.


Example 1:

Input: s = ["h","e","l","l","o"]
Output: ["o","l","l","e","h"]


Example 2:

Input: s = ["H","a","n","n","a","h"]
Output: ["h","a","n","n","a","H"]
 

Constraints:

1 <= s.length <= 10^5
s[i] is a printable ascii character.

```python
tests = [
	{
		"s": ["h","e","l","l","o"],
		"expectedOutput": ["o","l","l","e","h"]
	},
	{
		"s": ["H","a","n","n","a","h"],
		"expectedOutput": ["h","a","n","n","a","H"]
	},
]
```

# Solution #1
It has a time complexity of O(n) since it has to pass through every element of the string to rearrange them.
This works but by creating a new list, and we are asked to do it in the same given list.

```python
def reverse_string(s):
  """
  :type s: List[str]
  """
  result = []
  for i in range(len(s)):
    result.append(s[len(s) - i - 1])
  return result
```

# Solution #2
This function reverses the given list of characters `s` **in-place**. The in-place operation means that we do not create a new list but modify the original one.

We use two pointers: `left` starts from the beginning of the list, and `right` starts from the end.
In each iteration, we swap the elements at the `left` and `right` positions.
Both pointers move towards the center until they meet or cross.

Its time complexity is of O(n) because we traverse the entire list once with a single pass, swapping elements.

Its space complexity is of O(1) since this solution only uses a constant amount of extra space (the two pointers: `left` and `right`). An it modifies the list in-place, so no additional 
space is required.


This problem is more of a Space Complexity problem.

```python
def reverse_string(s):
  """
  :type s: List[str]
  """
  left, right = 0, len(s) - 1
  while left < right:
    s[left], s[right] = s[right], s[left]
    left +=1
    right -=1
  return s
```

```python
for test in tests:
  output = reverse_string(test["s"])
  print('output', output)
  print('>', output == test["expectedOutput"])
```
