# Container With Most Water

You are given an integer array heights where heights[i] represents the height of the ith bar.

You may choose any two bars to form a container. Return the maximum amount of water a container can store.

Example 1:

<img src="../assets/container-with-most-water-example-1.png" width="750">

```
Input: height = [1,7,2,5,4,7,3,6]
Output: 36
```

Example 2:

```
Input: height = [2,2,2]
Output: 4
```

Constraints:

```
2 <= height.length <= 1000
0 <= height[i] <= 1000
```

````python
tests = [
	{
		"heights": [1,7,2,5,4,7,3,6],
		"expectedOutput": 36
	},
	{
		"heights": [2,2,2],
		"expectedOutput": 4
	},
]
````

...

````python
def maxArea(heights):
    """
    use two pointers algorithm, set one pointer at the start and the other at the end
    iterate the array and calculate the waterCapacity on each pair (j - i) * min(height[i]m height[j])
    compare with latest maxWaterCapacity and update if its bigger
    iterate through the array by moving the pointer with the minumum height
    Time: O(n)
    Space: O(1)
    """
    maxWaterCapacity = 0
    i = 0
    j = len(heights) - 1
    while i != j:
        waterCapacity = (j - i) * min(heights[i], heights[j])
        maxWaterCapacity = max(maxWaterCapacity, waterCapacity)
        i, j = (i + 1, j) if heights[i] < heights[j] else (i, j - 1)
    return maxWaterCapacity
````

````python
def maxArea(heights):
    """
    Brute force
    Time: O(n^2)
    Space: O(1)
    """
    maxWaterCapacity = 0
    for i in range(len(heights)):
        for j in range(i + 1, len(heights)):
            waterCapacity = (j - i) * min(heights[i], heights[j])
            maxWaterCapacity = max(maxWaterCapacity, waterCapacity)
    return maxWaterCapacity
````

````python
for test in tests:
    output = maxArea(test["heights"])
    print('output',output)
    print('>', output == test["expectedOutput"])
````
